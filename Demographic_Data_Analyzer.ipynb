## <font color = "red"> Question 3 </font>

What is the percentage of people who have a Bachelor's degree?

----------

### Take a quick look at the dataset to know the name of the column required

```python
data.head()
```

#### To get percentage, two numbers are needed
#### <ul> <li> Total population of the dataset </li> <li> Population of individuals with Bachelors </li> </ul>

----------

#### First, we will begin with calculating total population of the dataset

NOTE: You can implement the "count" method on the whole dataset, but that will return an array of the same number, so to avoid that just pick one column only (can be any, but in our case we will pick the "education" column as its our focus)

```python
total_pop = data["education"].count()
print(total_pop)
```

#### Then, we will get the population of individuals with a bachelors

```python
bachelors_pop = data.loc[data["education"] == "Bachelors", "education"].count()
print(bachelors_pop)
```

## <font color = "red"> Solution to Question 3 </font>

#### Calculate percentage (round to nearest tenth as the instructions state)

```python
bachelors_percentage = round((bachelors_pop / total_pop) * 100, 1)
print(bachelors_percentage)
```

## <font color = "red"> Question 4 </font>

What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?

#### Firstly, we will need to narrow down the dataset to the population with an advanced education

```python
data.loc[(data["education"] == 'Bachelors') | (data["education"] == 'Masters') | (data["education"] == 'Doctorate')]
```

#### Now that we've gotten it, get the total amount of this population

```python
adv_edc_pop = data.loc[ (data["education"] == 'Bachelors') | (data["education"] == 'Masters') | (data["education"] == 'Doctorate'), "education"].count()
print(adv_edc_pop)
```

#### After that, we will need to narrow the dataset even more to both of our conditions (add another 'loc' method to add the other condition)

```python
data.loc[(data["education"] == 'Bachelors') | (data["education"] == 'Masters') | (data["education"] == 'Doctorate'), "salary"].loc[data["salary"] == ">50K"]
```

#### As we have narrowed down to our focused area, get the amount of this population

```python
adv_edc_greater50K_pop = data.loc[(data["education"] == 'Bachelors') | (data["education"] == 'Masters') | (data["education"] == 'Doctorate'), "salary"].loc[data["salary"] == ">50K"].count()
print(adv_edc_greater50K_pop)
```

## <font color = "red"> Solution to Question 4 </font>

#### Now we can calculate the percentage (round answer to the nearest tenth)

```python
adv_edc_greater50K_percentage = round(adv_edc_greater50K_pop / adv_edc_pop * 100, 1)
print(adv_edc_greater50K_percentage)
```

## <font color = "red"> Question 8 </font>

What country has the highest percentage of people that earn >50K and what is that percentage?

#### Lets take a quick look to get the name of the column we need

```python
data.head()
```

#### This question may seem to be complex at first, but good knowledge of pandas will reduce its complexity

#### Begin by narrowing the dataset to desired column, to do that we must also create the filter to get it narrowed down (obtain the countries with salary > 50K)

```python
filter = data["salary"] == ">50K"

data.loc[filter, "native-country"].to_frame()
```

#### Now we need to get individual counts of each country, to do this we can use a built-in pandas method: value_counts()

```python
data.loc[filter, "native-country"].value_counts()
```

#### To get percentage values, you can go multiple routes. However, the simplest would be to pass (normalize = True) in the value_counts method.... keep in mind though that it will be in decimal form, so multiply by 100 later on, to get percentage value

```python
countries_greater50K_percentages = data.loc[filter, "native-country"].value_counts(normalize = True)
countries_greater50K_percentages
```

#### We are not done yet, as the question asks for the maximum percentage... use the max method

```python
max_percentage_greater50K = countries_greater50K_percentages.max()

max_percentage_greater50K
```

## <font color = "red"> Question 9 </font>

Identify the most popular occupation for those who earn >50K in India.

#### Lets take a quick look at our dataset

```python
data.head()
```

#### There are 2 different methods we can use here

#### <ul> <li>Normal filtering method</li> <li>Groupby</li> </ul>

#### Lets begin with the filtering method we are used to, and solved most of the questions with

#### Narrow down to the 2 restrictions needed: Indian nationality & >50K salary

```python
data.loc[data["native-country"] == "India"].loc[data["salary"] == ">50K"]
```

#### Now lets access the occupation column

```python
data.loc[data["native-country"] == "India"].loc[data["salary"] == ">50K"]["occupation"]
```

## <font color = "red"> Solution to Question 9 </font>

#### Finally, apply the value_counts method to obtain the different values for occupation numbers

```python
data.loc[data["native-country"] == "India"].loc[data["salary"] == ">50K"]["occupation"].value_counts()
```

```python
data.loc[data["native-country"] == "India"].loc[data["salary"] == ">50K"]["occupation"].value_counts().max()
```

---------

#### Now lets get the solution to this question using our other technique (Groupby)

#### Begin by grouping the data set based on countries

```python
data.groupby(["native-country"])
```

#### After splitting the data into the required groups (countries), lets get the group we need (India)

```python
data.groupby(["native-country"]).get_group("India")
```

```python
data.groupby(["native-country"]).get_group("India")["salary"]
```
