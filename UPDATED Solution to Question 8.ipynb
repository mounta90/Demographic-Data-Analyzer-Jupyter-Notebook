## What country has the highest percentage of people that earn >50k and what is that percentage?

```python
countries_greater50K_count_series = df.loc[df["salary"] == ">50K", 'native-country'].value_counts()
```

```python
countries_greater50K_count_series
```

```python
countries_greater50K_percentage_series = (countries_greater50K_count_series / countries_count_series) * 100
```

```python
countries_greater50K_percentage_series_max = (countries_greater50K_count_series / countries_count_series).max()
```

```python
countries_greater50K_percentage_series_max
```

```python
countries_amount_series = df["native-country"].value_counts()

countries_greater50K_amount_series = df.loc[df["salary"] == ">50K", 'native-country'].value_counts()

countries_greater50K_percentage_series = (countries_greater50K_amount_series / countries_amount_series) * 100

countries_greater50K_percentage_series_max = countries_greater50K_percentage_series.max()

highest_earning_country_percentage = round(countries_greater50K_percentage_series_max, 1)

  
for i in range(countries_greater50K_percentage_series.size):
    print(f"Index Name: {countries_greater50K_percentage_series.index[i]} / Value: {countries_greater50K_percentage_series[i]}")
    if countries_greater50K_percentage_series[i] == countries_greater50K_percentage_series_max:
        highest_earning_country = countries_greater50K_percentage_series.index[i]
    else:
        pass
```
